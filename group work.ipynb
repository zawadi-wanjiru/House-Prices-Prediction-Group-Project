```python
import pandas as pd

# Load the dataset
file_path ='C:/Users/ADMIN/Downloads/house_prices_dataset.csv'
house_data = pd.read_csv(file_path)

# Display the first few rows and summary information
house_data.head(), house_data.info(), house_data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define numerical columns
numerical_cols = [
    "Size (sq. ft.)", 
    "Number of Rooms", 
    "Number of Bathrooms", 
    "Year Built", 
    "Distance to City Center (km)", 
    "Nearby Amenities", 
    "Price (USD)"
]

# Plot distributions for numerical columns
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(4, 2, i)
    sns.histplot(house_data[col], kde=True, color="blue", bins=15)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

```

```python
# Compute the correlation matrix
correlation_matrix = house_data[numerical_cols].corr()

# Plot a heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()

```

```python
# Boxplot of Price (USD) by Location
plt.figure(figsize=(8, 6))
sns.boxplot(x="Location", y="Price (USD)", data=house_data, palette="Set2")
plt.title("House Prices by Location")
plt.xlabel("Location")
plt.ylabel("Price (USD)")
plt.show()
```
